```python
from jarvis.db.figshare import data

dft3d = data("dft_3d")

print(len(dft3d))
```

```python
structure_lookup = {
    x["jid"]: x["atoms"]
    for x in dft3d
}
```

```python
import pandas as pd

ri_final = pd.read_parquet(
    "notebooks/data/processed/ri_final_no_density.parquet"
)

ri_final.shape
```

```python
jarvis = pd.read_parquet(
    "data/raw/jarvis_dft3d.parquet"
)

jarvis.shape
```

```python
import pandas as pd
from pymatgen.core import Composition

def to_reduced_formula(formula):
    try:
        return Composition(formula).reduced_formula
    except:
        return None

# make sure JARVIS has reduced_formula
jarvis["reduced_formula"] = jarvis["formula"].apply(to_reduced_formula)

# one representative JID per reduced formula
jarvis_rf = (
    jarvis[["reduced_formula", "jid"]]
    .dropna(subset=["reduced_formula"])
    .drop_duplicates(subset=["reduced_formula"], keep="first")
)

# merge using reduced_formula, not formula
ri_structure = ri_final.merge(
    jarvis_rf,
    on="reduced_formula",
    how="left"
)

ri_structure["atoms"] = ri_structure["jid"].map(structure_lookup)

print(ri_structure.shape)
print(ri_structure["atoms"].notna().sum())
```

```python
ri_structure["atoms"].notna().mean()
```

```python
ri_structure.to_parquet(
    "notebooks/data/processed/ri_structure_dataset.parquet",
    index=False
)
```
